```python
# Download dataset directly to Unity Catalog Volume
import opendatasets as od

# Define Unity Catalog Volume path
volume_path = "/Volumes/workspace/default/ecommerce_data"

# Download directly to volume (opendatasets will create subdirectory)
od.download(
    "https://www.kaggle.com/datasets/ihasan88/customer-behavior-and-churn-prediction-dataset", 
    volume_path
)

print(f"Dataset downloaded to: {volume_path}")
```

```python
# Read datasets into PySpark DataFrame
df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Volumes/workspace/default/ecommerce_data/customer-behavior-and-churn-prediction-dataset/ecommerce_customer_data.csv")
```
